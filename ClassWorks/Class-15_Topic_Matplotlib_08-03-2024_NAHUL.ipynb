```python
#lineplot

import matplotlib.pyplot as plt
x=list(range(0,100,10))  #start 0, end 100, interval 10
y=list(range(0,50,5))  #start 0, end 50, interval 5
print(x)
print(y)
plt.plot(x,y,color='red',linewidth=1.0,linestyle=":",alpha=1, marker='x') #marker='o'
plt.xlabel("range 1-100")
plt.ylabel("range 1-50")
plt.title("Get line plot")
plt.show()
```

```python
#lineplot

import matplotlib.pyplot as plt
x=[3,8,6,10]
y=[6,8,4,6]
z=[6,9,12,15]
print(x)
print(y)
print(z)
plt.plot(x,y,color='red')
plt.plot(x,z, color='green')
plt.title("Multiple plots")
plt.legend(['a','b'])
plt.show()
```

```python
#lineplot subplot usages

import matplotlib.pyplot as plt
x=[3,8,6,10]
y=[6,8,4,6]
z=[6,9,12,15]

print(x)
print(y)
print(z)

plt.subplot(2,2,1)
plt.plot(x,y,color='red')
plt.title("x vs y")
plt.legend(['a'])

plt.subplot(2,2,2)
plt.plot(x,z, color='green')
plt.title("x vs z")
plt.legend(['b'])


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x=np.arange(0,10,1)
y1=2*x+3
y2=4*x**2+2
y3=x**1/2

plt.subplot(3,1,1)
plt.plot(x,y1,color='red',linewidth=1.0,linestyle=":",alpha=1, marker='o')
plt.title("graph 1")
plt.legend(['lines'],loc='best')
plt.grid()

plt.subplot(3,1,2)
plt.plot(x,y2,color='green',linewidth=1.0,linestyle=":",alpha=1, marker='x')
plt.title("graph 2")
plt.legend(['lines'],loc='best')
plt.grid()

plt.subplot(3,1,3)
plt.plot(x,y3,color='blue',linewidth=1.0,linestyle=":",alpha=1, marker='o')
plt.title("graph 3")
plt.legend(['lines'],loc='best')
plt.grid()

plt.show()
```

```python
#scatterplot

import matplotlib.pyplot as plt
x=[3,8,6,10]
y=[6,8,4,6]
print(x)
print(y)
plt.scatter(x,y,color='red')
plt.scatter(x,y,color='green',s=1000,edgecolor='blue',alpha=1, marker='4')
plt.title("Get scatter plot")
plt.show()
```

```python
#barplot

import matplotlib.pyplot as plt
x=['mango','guava','banana']
y=[4,8,2]
plt.bar(x,y,color='green')
plt.xlabel("fruits")
plt.ylabel("no of fruits")
plt.title("fruit distribution")
plt.show()
```

```python
#barplot

import matplotlib.pyplot as plt
x=['mango','guava','banana']
y=[4,8,2]
plt.barh(x,y,color='green')
plt.xlabel("fruits")
plt.ylabel("no of fruits")
plt.title("fruit distribution")
plt.show()
```

```python
 import matplotlib.pyplot as plt
%matplotlib inline

label=['a','b','c','d']
size=[25,40,15,20]

plt.pie(size,labels=label,autopct= '%1.3f%%', shadow=True, startangle=90, explode=(0,0,0,0))
plt.title("pie chart demo")
plt.show()

"""
autopct='%1.3f%%' -> the percentage in pie chart will be shown in decimal after three digits, 1.2 will be two digits
shadow=True       -> shadows will appear in slices
startangle=90     -> the main pie will start from 90, mif 100, 20, 60 etc angle can be put
explode=(0,0,0,0) -> the slices will get out from the chart according to i/p (0,0.2,0,0.5)  """
```
