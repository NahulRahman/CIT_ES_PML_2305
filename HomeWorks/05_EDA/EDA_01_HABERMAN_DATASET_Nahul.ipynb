# EXPLORATORY DATA ANALYSIS : HABERMAN DATASET

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv("C:\\Users\\Acer\\ES PML 2305\\HW\\haberman.csv",names=['age', 'operation_year', 'axil_nodes', 'survival_status'])
df
```

```python
df['survival_status'].replace([1,2], ["yes","No"], inplace = True)
df
```

```python
df['survival_status'].value_counts()
```

```python
df['survival_status'].unique()
```

```python
yes = df.loc[df["survival_status"]=="yes"]
no = df.loc[df["survival_status"]=="No"]
```

```python
print("Describing the dataset of patients who survived or YES\n")
print(yes.describe())
print("\n\n\n")
print("Describing the dataset of patients who didn't survived or NO\n")
print(no.describe())
print("SURVIVAL STATUS : YES\n\nObservations :\n\nNumber of people survived : 225\naverage age of people that survived : 52\nminimum age of people that survived : 30\maximum age of people that survived : 77\n\nSURVIVAL STATUS : NO\nObservations :\nNumber of people who did not survive : 81\naverage age of people who did not survive : 53\nminimum age of people who did not survive : 34\nmaximum age of people who did not survive : 83\n")
```

```python

sns.boxplot(x = "survival_status", y = "age", data  = df)
plt.show()

sns.boxplot(x = "survival_status", y = "operation_year", data  = df)
plt.show()

sns.boxplot(x = "survival_status", y = "axil_nodes", data  = df)
plt.show()
```

```python
sns.violinplot(x = "survival_status", y = "axil_nodes", data  = df)
plt.show()

sns.violinplot(x = "survival_status", y = "operation_year", data  = df)
plt.show()

sns.violinplot(x = "survival_status", y = "age", data  = df)
plt.show()
```

```python
df.plot(kind ="scatter", x="operation_year", y = "age")
plt.grid()
plt.show()

sns.set_style("whitegrid");
sns.FacetGrid(df, hue="survival_status", height=8).map(plt.scatter, "age", "operation_year").add_legend();
plt.show();

sns.set_style("whitegrid");
sns.FacetGrid(df, hue="survival_status", height=8) \
   .map(plt.scatter, "axil_nodes", "operation_year") \
   .add_legend();
plt.show();
```

```python
sns.set_style("whitegrid");
sns.pairplot(df,hue="survival_status",height=6)
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df,hue="survival_status",height=6).map(sns.distplot,"age").add_legend();
plt.title('Histogram of ages of patients', fontsize=17)
plt.show();

sns.FacetGrid(df,hue="survival_status",height=6).map(sns.distplot,"operation_year").add_legend();
plt.title('Histogram of operation year of patients', fontsize=17)
plt.show();


sns.FacetGrid(df,hue="survival_status",height=6).map(sns.distplot,"axil_nodes").add_legend();
plt.title('Histogram of auxillary nodes detected', fontsize=17)
plt.show();
```

```python
sns.jointplot(x="age",y="operation_year",data=df, kind="kde")
plt.show()

sns.jointplot(x="age",y="axil_nodes",data=df, kind="kde")
plt.show()

sns.jointplot(x="operation_year",y="axil_nodes",data=df, kind="kde")
plt.show()
```

```python
df['survival_status'].value_counts().plot(kind= "bar", color= 'blue' ,figsize =(8,6))
plt.xlabel("number of survivals", labelpad = 8)
plt.ylabel("survival_count", labelpad = 8)
plt.title("Survival Status", y = 1.02)
plt.legend()
plt.grid()
plt.show()
```

```python
df['survival_status'].value_counts().plot(kind= "pie",autopct = '%1.2f%%', shadow= True, startangle = 90, explode = (0,0))
plt.title("pie Chart Demo")
plt.show()
```
